## Merging or Join


```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
df1
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
df2
```

```python
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
df3
```

```python
frames = [df1, df2, df3]
frames
```

```python
result = pd.concat(frames)
result
```

```python
df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[2, 3, 6, 7],
)
df4
```

```python
result=pd.concat([df1,df4],axis=1)
result
```

```python
print(df1)
print(df4)
```

```python
result=pd.concat([df1,df4],ignore_index=True,sort=False)
result
```

```python
df_right=pd.merge(df1,df2,left_on='A',right_on='A',how='left')
df_right
```

```python
df_right=pd.merge(df1,df2,left_on='A',right_on='A',how='right')
df_right
```

## pandas.concat()

```python
s1=pd.Series(['a','b','c'])
s2=pd.Series(['d','e','f'])
sample=pd.concat([s1,s2])
sample
```

```python
sample=pd.concat([s1,s2],keys=['s1','s2'],names=['key','index'])
sample
sample=pd.DataFrame(sample,columns=['values'])
sample
```
