### Making Changes to the data

```python
data['Total']=data.iloc[:,4:10].sum(axis=1)
data
```

```python
cols=list(data.columns)
data=data[cols[0:10]+[cols[-1]]
          +cols[10:12]]
data
```

```python
# Dropping columns by index positions
columns_to_drop = [12,13,14,15,16,17,18 ]
new_data = data.drop(data.columns[columns_to_drop], axis=1)

print("\nDataset after dropping columns by index positions:")
display(new_data)
```

```python
new_data['Total']=new_data.iloc[:,4:10].sum(axis=1)
new_data
```

```python
cols=list(data.columns)
new_data=new_data[cols[0:10]+[cols[-1]]
          +cols[10:11]]
```
