### Filtering Data

```python
import pandas as pd
```

```python
data=pd.read_csv('modified.csv')
```

```python
data.head(3)
```

```python
new_df=data.loc[(data['Type 1']=='Grass') & (data['Type 2']=='Poison')]
```

```python
new_df
```

```python
new_df.reset_index(drop=True,inplace=True)
new_df
```

```python
mega_df=data.loc[~data['Name'].str.contains('Mega')]  # ~ - it is used for 'not' operator 
mega_df.reset_index(drop=True,inplace=True)
mega_df
```

## Use Regex

```python
# using regex
import re
data.loc[data['Name'].str.contains('^Bu[a-z]*y$',flags=re.I,regex=True)]  # it is find name starts with Bu and having multiple charecter and end with r
```

## Aggregate Statistics (Group By)

```python
import numpy as np
data1=pd.read_csv('modified.csv')
```

```python
data1.head()
```

```python
data1.groupby(['Type 1'])['Attack'].min()
```

```python
data1.groupby(['Type 1'])['Attack'].max()
```

```python
all_count=data1.groupby(['Type 1','Type 2']).count()
```
